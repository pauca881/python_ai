```python
import pygame
import random
import sys

# Inicializar pygame
pygame.init()

# Definir dimensiones de la pantalla
ANCHO, ALTO = 800, 600
PANTALLA = pygame.display.set_mode((ANCHO, ALTO))
pygame.display.set_caption("Juego de evitar los enemigos")

# Definir colores
NEGRO = (0, 0, 0)

# Configuración de la nave
NAVE_ANCHO, NAVE_ALTO = 50, 50
nave_img = pygame.image.load('images/htlr.png')
nave_img = pygame.transform.scale(nave_img, (NAVE_ANCHO, NAVE_ALTO))
nave_x, nave_y = ANCHO // 2, ALTO - 100
VELOCIDAD_NAVE = 5

# Configuración de los asteroides
ASTEROIDE_ANCHO, ASTEROIDE_ALTO = 50, 50
asteroide_img = pygame.image.load('images/jw.png')
asteroide_img = pygame.transform.scale(asteroide_img, (ASTEROIDE_ANCHO, ASTEROIDE_ALTO))
VEL_ASTEROIDE = 5
asteroides = []

# Crear un asteroide nuevo
def crear_asteroide():
    x = random.randint(0, ANCHO - ASTEROIDE_ANCHO)
    y = -ASTEROIDE_ALTO
    asteroides.append(pygame.Rect(x, y, ASTEROIDE_ANCHO, ASTEROIDE_ALTO))

# Actualizar la posición de los asteroides
def mover_asteroides():
    for asteroide in asteroides:
        asteroide.move_ip(0, VEL_ASTEROIDE)

# Eliminar asteroides que ya salieron de la pantalla
def eliminar_asteroides_fuera():
    global asteroides
    asteroides = [ast for ast in asteroides if ast.top < ALTO]

# Verificar si hay colisión entre la nave y algún asteroide
def check_colision():
    nave_rect = pygame.Rect(nave_x, nave_y, NAVE_ANCHO, NAVE_ALTO)
    for asteroide in asteroides:
        if nave_rect.colliderect(asteroide):
            return True
    return False

# Bucle principal
reloj = pygame.time.Clock()
corriendo = True
contador = 0
while corriendo:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Controlar la nave con las teclas
    teclas = pygame.key.get_pressed()
    if teclas[pygame.K_LEFT]:
        nave_x -= VELOCIDAD_NAVE
    if teclas[pygame.K_RIGHT]:
        nave_x += VELOCIDAD_NAVE
    if teclas[pygame.K_UP]:
        nave_y -= VELOCIDAD_NAVE
    if teclas[pygame.K_DOWN]:
        nave_y += VELOCIDAD_NAVE

    # Limitar el movimiento de la nave dentro de la pantalla
    nave_x = max(0, min(nave_x, ANCHO - NAVE_ANCHO))
    nave_y = max(0, min(nave_y, ALTO - NAVE_ALTO))

    # Cada 30 cuadros, crear un nuevo asteroide
    if contador % 30 == 0:
        crear_asteroide()

    # Mover los asteroides y eliminar los que están fuera
    mover_asteroides()
    eliminar_asteroides_fuera()

    # Verificar colisión entre la nave y un asteroide
    if check_colision():
        corriendo = False

    # Dibujar la pantalla
    PANTALLA.fill(NEGRO)
    PANTALLA.blit(nave_img, (nave_x, nave_y))
    for asteroide in asteroides:
        PANTALLA.blit(asteroide_img, (asteroide.left, asteroide.top))

    pygame.display.flip()
    reloj.tick(60)
    contador += 1

print("¡Has perdido! Gracias por jugar.")
```
