```python
# Realiza un análisis de correlación para evaluar la relación entre las variables.
# ¿Cuáles parecen estar más relacionadas con los ingresos futuros y con la
# rentabilidad?
```

```python
# Análisis de correlación

# Calcular la matriz de correlación
correlation_matrix = df.corr()
print(correlation_matrix)

#Visualizar la matriz de correlación
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Visualiza las distribuciones de edad, ingreso_anual, saldo_promedio,
# gastos_mensuales y historial_crediticio
```

```python
# Visualizar distribuciones
variables = ['edad', 'ingreso_anual', 'saldo_promedio', 'gastos_mensuales', 'historial_crediticio']

plt.figure(figsize=(15, 10))
for i, variable in enumerate(variables, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[variable], kde=True)
    plt.title(f'Distribución de {variable}')

    
plt.tight_layout()
plt.show()
```

```python
# Gráficos de dispersión
plt.figure(figsize=(15, 5))

# Ingreso anual vs Ingresos futuros
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='ingreso_anual', y='ingresos_futuros')
plt.title('Ingreso Anual vs Ingresos Futuros')

# Saldo promedio vs Ingresos futuros
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='saldo_promedio', y='ingresos_futuros')
plt.title('Saldo Promedio vs Ingresos Futuros')

# Ingreso anual vs Saldo promedio
plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='ingreso_anual', y='saldo_promedio')
plt.title('Ingreso Anual vs Saldo Promedio')

plt.tight_layout()
plt.show()
```
