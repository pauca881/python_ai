```python
import pandas as pd

csv = pd.read_csv('employees_extra.csv')
excel1 = pd.read_excel('employees_basic.xlsx')
excel2 = pd.read_excel('employees_addresses.xlsx')
json = pd.read_json('employees_salaries.json')

csv.drop_duplicates(subset='ID', inplace=True)
excel1.drop_duplicates(subset='ID', inplace=True)
excel2.drop_duplicates(subset='ID', inplace=True)
excel2.drop_duplicates(subset='ID', inplace=True)

dataframe = pd.merge(excel1, json, on='ID', how='inner')
dataframe = pd.merge(dataframe, excel2, on='ID', how='left')

valores_reemplazo = {
    'salario': 0,
    'direccion': 'N/A',
    'ciudad': 'N/A',
    'país': 'N/A'
}

dataframe.fillna(valores_reemplazo)

dataframe_final = pd.concat([dataframe, csv], axis=0)
dataframe_final.to_csv('employees_final.csv')
dataframe_final
```
