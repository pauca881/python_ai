```python
# Crea una figura con dos subplots en una columna: un gráfico de líneas que muestre los
# valores [10, 15, 20, 25, 30] frente a los días [1, 2, 3, 4, 5], y un gráfico de
# barras que muestre las ventas de tres productos (['A', 'B', 'C']) con los valores
# [200, 150, 100]. Añade títulos a ambos gráficos.
```

```python
import matplotlib.pyplot as plt

# Datos
dias = [1, 2, 3, 4, 5]
valores_lineas = [10, 15, 20, 25, 30]
productos = ['A', 'B', 'C']
ventas = [200, 150, 100]

#creacio de la figura i els subplots
fig, axs = plt.subplots(2, 1, figsize=(6, 8))

#gràfic de línees
axs[0].plot(dias, valores_lineas, marker='o', color='b', linestyle='-', linewidth=2)
axs[0].set_title('Valores frente a días')
axs[0].set_xlabel('Días')
axs[0].set_ylabel('Valores')

#gràfic de barres
axs[1].bar(productos, ventas, color=['green', 'orange', 'red'])
axs[1].set_title('Ventas de productos')
axs[1].set_xlabel('Productos')
axs[1].set_ylabel('Ventas')


plt.tight_layout()

# Mostrar la figura
plt.show()
```

```python
# Usando Seaborn, crea un gráfico que muestre tanto el histograma como la densidad de
# probabilidad estimada (KDE) de un conjunto de datos generados aleatoriamente que sigan
# una distribución normal. Genera 1000 datos aleatorios con una media de 0 y una desviación
# estándar de 1.
```

```python
import numpy as np

# Generar datos aleatorios con distribución normal
data = np.random.normal(loc=0, scale=1, size=1000)

#aqui creo el gráfico
sns.histplot(data, kde=True, color='blue', bins=30)
plt.title('Histograma y KDE de datos generados de manera random')

plt.show()
```

```python
# Utilizando Seaborn, crea un mapa de calor que muestre la correlación entre las variables de
# un DataFrame de ejemplo. Crea un DataFrame con las columnas edad, ingresos y
# ahorros, y muestra su matriz de correlación usando un mapa de calor con anotaciones.
```

```python
data = pd.DataFrame({
'edad': [22, 25, 30, 35, 40, 45, 50],
'ingresos': [2000, 2400, 3000, 3500, 4000, 4500, 5000],
'ahorros': [1000, 1200, 1500, 1800, 2000, 2500, 3000]
})

#  matriz de correlación
correlation_matrix = data.corr()

#mapa de calor con Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Mapa de calor de la correlación entre Edad, Ingresos y Ahorros')
```

```python
# Crea un gráfico de líneas interactivo usando Plotly que muestre la evolución de la
# temperatura diaria durante una semana. Los días de la semana son ['Lunes',
# 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'] y las
# temperaturas en grados Celsius son [20, 21, 19, 22, 23, 20, 18]. Añade títulos y
# etiquetas a los ejes.
```

```python
import plotly.express as px

data = pd.DataFrame({
'Días': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes',
'Sábado', 'Domingo'],
'Temperatura': [20, 21, 19, 22, 23, 20, 18]
})

#gráfico de líneas interactivo
fig = px.line(data, x='Días', y='Temperatura', title='Evolución de la Temperatura Durante la Semana')

#etiquetas a los ejes
fig.update_layout(
    xaxis_title='Días de la Semana',
    yaxis_title='Temperatura (°C)',
    title_x=0.5  # Centrar el título
)
```

```python
# Crea un gráfico de barras interactivo que muestre las ventas de tres productos diferentes
# durante un mes. Los productos son ['Producto A', 'Producto B', 'Producto
# C'], y las ventas respectivas son [300, 400, 500]. Añade un título al gráfico y etiquetas
# a los ejes.
```

```python
data = pd.DataFrame({
'Producto': ['Producto A', 'Producto B', 'Producto C'],
'Ventas': [300, 400, 500]
})

#creación gráfico
fig = px.bar(data, 
             x='Producto', 
             y='Ventas',
             title='Ventas Mensuales por Producto',
             labels={'Producto': 'Tipo de Producto',
                    'Ventas': 'Cantidad de Ventas'},
             color='Producto',
             text='Ventas')

fig.show()
```

```python
# Crea un gráfico de dispersión interactivo que muestre la relación entre el peso y la altura de
# un grupo de personas. Genera un DataFrame con las columnas peso, altura y edad, y
# # utiliza la columna edad para variar el tamaño de los puntos en el gráfico.
```

```python
data = pd.DataFrame({
'Peso': [70, 80, 90, 60, 65, 75, 85],
'Altura': [1.75, 1.80, 1.85, 1.65, 1.70, 1.78, 1.82],
'Edad': [25, 30, 35, 40, 45, 50, 55]
})

# Crear el gráfico de dispersión interactivo
fig = px.scatter(data, 
                x='Peso', 
                y='Altura',
                size='Edad',  #eel tamaño de los puntos varía según la edad
                color='Edad', #el color de los puntos varía según la edad
                title='Relación entre peso y altura según edad',
                labels={'Peso': 'Peso (kg)',
                       'Altura': 'Altura (m)',
                       'Edad': 'Edad (años)'})  

fig.show()
```
