```python
import folium

# Crear el mapa centrado en Barcelona
mapa = folium.Map(location=[41.3851, 2.1734], zoom_start=12)

# Definir la función de estilo según densidad de población
def estilo_barcelona(feature):
    density = feature['properties'].get('density', 0)
    if density > 1000:
        fillColor = 'red'
    elif density > 500:
        fillColor = 'orange'
    else:
        fillColor = 'green'
    
    return {
        'fillColor': fillColor,
        'color': 'black',
        'weight': 1.5,
        'fillOpacity': 0.7
    }

# Añadir archivo GeoJSON de Barcelona
folium.GeoJson(
    "barcelona.geojson",  # Archivo con los datos geográficos de Barcelona
    style_function=estilo_barcelona,
    tooltip=folium.GeoJsonTooltip(
        fields=["name", "density"],  # Campos a mostrar
        aliases=["Nombre del Distrito", "Densidad de Población"],  # Alias para los campos
        localize=True
    ),
    popup=folium.GeoJsonPopup(
        fields=["name", "density"], 
        aliases=["Nombre del Distrito", "Densidad de Población"]
    )
).add_to(mapa)

# Mostrar el mapa
mapa
```

```python
import pandas as pd
import folium

# Cargar el archivo GeoJSON de Barcelona y el dataframe con los datos
geo_json = "barcelona.geojson"
df = pd.DataFrame({
    'district': ['Eixample', 'Ciutat Vella', 'Sants-Montjuïc'],
    'density': [800, 1200, 600]
})

# Crear el mapa
mapa = folium.Map(location=[41.3851, 2.1734], zoom_start=12)

# Añadir el coroplético
folium.Choropleth(
    geo_data=geo_json,
    data=df,
    columns=['district', 'density'],
    key_on='feature.properties.name',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Densidad de población',
).add_to(mapa)

# Mostrar el mapa
mapa
```

```python
from folium import PolyLine
from folium.plugins import PolyLineTextPath

# Crear el mapa
mapa = folium.Map(location=[41.3851, 2.1734], zoom_start=12)

# Coordenadas de una ruta de ejemplo
ruta = [[41.3809, 2.1672], [41.3874, 2.1686], [41.3891, 2.1734]]

# Añadir la polilínea
polyline = PolyLine(locations=ruta, color="blue", weight=6)
mapa.add_child(polyline)

# Decorar la línea con flechas
PolyLineTextPath(
    polyline,
    '→ ',  # Texto para mostrar (las flechas)
    repeat=True,
    offset=10,
    attributes={'fill': 'red', 'font-weight': 'bold', 'font-size': '24'}
).add_to(mapa)

# Mostrar el mapa
mapa
```

```python
import folium
import matplotlib.pyplot as plt
from PIL import Image
import base64
from io import BytesIO

# Crear el gráfico con matplotlib
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
ax.set_title("Gráfico de Ejemplo")

# Guardar el gráfico como imagen en formato base64
buf = BytesIO()
fig.savefig(buf, format="png")
data = base64.b64encode(buf.getbuffer()).decode("ascii")

# Crear el mapa
mapa = folium.Map(location=[41.3851, 2.1734], zoom_start=12)

# Insertar el gráfico en un popup de folium
html = f'<img src="data:image/png;base64,{data}">'
iframe = folium.IFrame(html, width=300, height=300)
popup = folium.Popup(iframe, max_width=300)
folium.Marker([41.3851, 2.1734], popup=popup).add_to(mapa)

# Mostrar el mapa
mapa
```
