```python
#SCRAPPINNG
#pip install pyautogui
# crear un script básico  de automatización (os, sys, pyautogui)...de algún proceso.
import pyautogui
import time
import os

#24hrs en segundos
intervalo = 24 * 60 * 60  

def iniciar_sesion_wallapop():

    os.system("start chrome https://es.wallapop.com/auth/signin?redirectUrl=%2F")
    time.sleep(5)  # Espera a que se cargue la página

    #primero hago clic el botón de escribir iniciar sesión
    pyautogui.click(x=1000, y=100)  #me falta ajustar las coordenadas
    time.sleep(2)

    pyautogui.typewrite("micorreo@mail.com")
    pyautogui.press('tab')
    time.sleep(1)

    #escribo passsword
    pyautogui.typewrite("mi_password")
    time.sleep(1)

    #clic en el botón de iniciar sesión
    pyautogui.click(x=1000, y=400)#debo ajustar estas coordenadas
    time.sleep(5)  #espero a que se cargue la página principal

def eliminar_y_subir_producto():

    #primero incnio sesión en Wallapop
    iniciar_sesion_wallapop()

    #espero otra vez un momento para que la interfaz cargue
    time.sleep(5)

    #abro la pestaña del producto en Wallapop (ajusto las coordenadas)
    #m aseguro de que el producto esté visible en pantalla
    pyautogui.click(x=100, y=200)  #coordenadas del producto en tu pantalla
    time.sleep(2)

    # Hacer clic en "Eliminar"
    pyautogui.click(x=150, y=250)  # Coordenadas del botón "Eliminar"
    time.sleep(1)

    # Confirmar la eliminación
    pyautogui.click(x=200, y=300)  # Coordenadas de "Confirmar" eliminación
    time.sleep(2)

    #creo un nuevo anuncio (ajusta las coordenadas)
    pyautogui.click(x=50, y=50)  #coord del botón "Añadir anuncio"
    time.sleep(2)

    #relleno la info del producto (ajusto las coordenadas y el texto)
    pyautogui.typewrite("Iphone 15")
    pyautogui.typewrite(["tab"])
    pyautogui.typewrite("Estado: Nuevo")
    pyautogui.typewrite(["tab"])
    pyautogui.typewrite("Precio: 1000")  #precio producto
    time.sleep(2)

    #hago clic en "Publicar"
    pyautogui.click(x=300, y=350)  #coord del botón "Publicar"
    time.sleep(2)

#bucle para automatizar el proceso cada 24 horas
while True:
    eliminar_y_subir_producto()
    time.sleep(intervalo)


```
