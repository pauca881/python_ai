```python
#Herencia
class Coche:
    ruedas = 4
    #Si pusiese ruedas en el constructor, se podria modificar
    
    def __init__(self, color, acceleracion):
        self.color = color
        self.acceleracion = acceleracion
        self.velocidad = 0
        #Velocidad se puede modificar, pero siempre estará de inicio 0, no se puede modificar al crear
        
    def acelerar(self):
        self.velocidad = self.velocidad + self.acceleracion

    def frenar(self):
        v = self.velocidad - self.acceleracion
        if v < 0:
            v = 0
            self.velocidad = v

class CocheVolador(Coche):
    ruedas = 6

    def __init__(self, color, acceleracion, esta_volando):
        #super, se refiere al padre es decir coche.init
        super().__init__(color, acceleracion)
        self.esta_volando = esta_volando
        
    def volar(self):
        self.esta_volando = True

    def aterrizar(self):
        self.esta_volando = False


seat = Coche('rojo', 100)
seat.acelerar()
seat.velocidad
seat.acelerar()
seat.velocidad
seat.acelerar()
seat.velocidad
seat.acelerar()
seat.velocidad
seat.frenar()
seat.velocidad

delorean = CocheVolador('azul', 110, False)
seat_volador = CocheVolador('rojo', 100, False)

#return True if seat pertains to a Coche
isinstance(seat,Coche)

#return True if CocheVolador is a subclass of Coche
issubclass(CocheVolador, Coche)
```

```python
class Libro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor
        self.disponible = True

    def prestar(self):
        if self.disponible:
            self.disponible = False
            print(f'El libro {self.titulo} ha sido prestado')
        else:
            print(f'El libro {self.titulo} no esta disponible para ser prestado')

    def devolver(self):
        if not self.disponible:
            self.disponible = True
            print(f'El libro {self.titulo} ha sido devuelto')
        else: 
            print(f'El libro {self.titulo} no se puede devolver porque ya esta disponible')



    def mostrar_informacion(self):
        estado = "Disponible" if self.disponible else "No disponible"
        print(f"Título: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Estado: {estado}")
        print()

    

class Biblioteca:
    def __init__(self, nombre):
        self.nombre = nombre
        self.libros_disponibles = []
    def agregar_libro(self, libro):
        self.libros_disponibles.append(libro)
        print(f"El libro '{libro.titulo}' ha sido agregado a la biblioteca.")
    def mostrar_libros(self):
        if not self.libros_disponibles:
            print("No hay libros disponibles en la biblioteca.")
        else:
            print("Libros disponibles en la biblioteca:")
            for libro in self.libros_disponibles:
                libro.mostrar_informacion()

   
# Creación de instancias de las clases
libro1 = Libro("El principito", "Antoine de Saint-Exupéry")
libro2 = Libro("1984", "George Orwell")
libro3 = Libro("Don Quijote de la Mancha", "Miguel de Cervantes")

biblioteca = Biblioteca("Biblioteca Central")
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)
biblioteca.agregar_libro(libro3)

libro1.prestar()
libro2.prestar()
biblioteca.mostrar_libros()
libro1.devolver()
biblioteca.mostrar_libros()
```

```python
#Ejercicio 1
class Persona:
    def __init__(self, nombre, edad, DNI):
        self.nombre = nombre
        self.edad = edad
        self.DNI = DNI

#Codigo más legible y seguro
# persona = Persona("Juan", 30, "12345678A")
# print(persona.nombre)  # Acceso al getter
# persona.edad = 31      # Llama al setter
# persona.edad = -5    # Esto lanzaría un ValueError

    #@property sirva para que pueda hacer persona.nombre i no persona.get_nombre()

    # Getter para el nombre
    @property
    def nombre(self):
        return self._nombre

    # Setter para el nombre
    @nombre.setter
    def nombre(self, nombre):
        self._nombre = nombre

    # Getter para la edad
    @property
    def edad(self):
        return self._edad

    # Setter para la edad
    @edad.setter
    def edad(self, edad):
        if edad < 0:
            raise ValueError("La edad no puede ser negativa.")
        self._edad = edad

    # Getter para el DNI
    @property
    def DNI(self):
        return self._DNI

    # Setter para el DNI
    @DNI.setter
    def DNI(self, DNI):
        self._DNI = DNI   


    def mostrar(self):
        print(f"Nombre: {self.nombre}")
        print(f"Edad: {self.edad}")
        print(f"DNI: {self.DNI}")
        print()
        
    def esMayorDeEdad(self):
        mayor_de_edad = False

        if(self.edad >= 18 ):
            print('El cliente es mayor de edad')
        else: print('El cliente es menor de edad')

p1 = Persona('Pau', 21, '49486A')
p2 = Persona('Mireia', 20, '49486A')

p1.mostrar()
p1.esMayorDeEdad()
```

```python
#Ejercicio 2 y 3
class Cuenta():
    
    def __init__(self, titular, cantidad=0.0):
        self.titular = titular
        self.cantidad = cantidad

    @property
    def titular(self):
        return self.titular

    # Setter para el nombre
    @titular.setter
    def titular(self, titular):
        self._titular = titular


    @property
    def cantidad(self):
        return self.cantidad

    # Setter para el nombre
    @cantidad.setter
    def cantidad(self, cantidad):
        self._cantidad = cantidad

    def mostrar(self):
        print(f"Titular: {self.titular}")
        print(f"Dinero: {self.cantidad}")
        print()

    def ingresar_cantidad(self, cantidad):
        if cantidad > 0:
            self._cantidad += cantidad
            print(f'Has ingresado {cantidad}€')
            print(f'Tienes un saldo de {self._cantidad}€')
            print()

        else:
            raise ValueError('ERROR: La cantidad debe ser positiva')
        
    def retirar_cantidad(self, cantidad):
        if cantidad > 0 and cantidad <= self._cantidad:
            self._cantidad -= cantidad
            print(f'Has retirado {cantidad}€')
            print(f'Tienes un saldo de {self._cantidad}€')
            print()

        else:
            raise ValueError('ERROR: La cantidad debe ser positiva o que haya suficiente')
        

c1 = Cuenta('Pau',20000)
c2 = Cuenta('Mireia')

c1.ingresar_cantidad(2500)
c1.retirar_cantidad(1000)


class CuentaJoven(Cuenta)
    
    def __init__(self, titular, cantidad=0.0, bonificacion=0):
        super().__init__(titular, cantidad)
        self.bonificacion = bonificacion

    @property
    def bonificacion(self):
        return self._bonificacion

    @bonificacion.setter
    def bonificacion(self, bonificacion):
        self._bonificacion = bonificacion

    def esTitularValido(self):
        return 18 <= self.edad < 25

    def retirar_dinero(self, cantidad):
        if self.esTitularValido():
            super().retirar_dinero(cantidad)
        else:
            raise ValueError("El titular no es válido para realizar la retirada de dinero.")

    def mostrar(self):
        return f"Cuenta Joven: Bonificación {self.bonificacion}%"
```
