```python
import random
import time
from datetime import datetime
from functools import wraps


def hangman():
# Palabras para el juego
    palabras = ["python", "programacion", "ordenador", "juego", "ahorcado"]
       

    # Decoradores
    def registrar_tiempo(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            inicio = time.time()
            resultado = func(*args, **kwargs)
            tiempo_total = time.time() - inicio
            with open("log_tiempo_juego.txt", "a") as f:
                f.write(f"Tiempo de juego: {tiempo_total:.2f} segundos\n")
            return resultado
        return wrapper

    def contador_juegos(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            wrapper.contador += 1
            with open("log_contador_juegos.txt", "w") as f:
                f.write(f"Número de juegos jugados: {wrapper.contador}\n")
            return func(*args, **kwargs)
        wrapper.contador = 0
        return wrapper

    def registrar_fecha_hora(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            ahora = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            with open("log_fecha_hora.txt", "a") as f:
                f.write(f"Partida jugada el: {ahora}\n")
            return func(*args, **kwargs)
        return wrapper

    def contador_victorias(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            resultado = func(*args, **kwargs)
            if resultado == "victoria":
                wrapper.victorias += 1
                print(f"Victorias totales: {wrapper.victorias}")
            return resultado
        wrapper.victorias = 0
        return wrapper

    def contador_derrotas(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            resultado = func(*args, **kwargs)
            if resultado == "derrota":
                wrapper.derrotas += 1
                print(f"Derrotas totales: {wrapper.derrotas}")
            return resultado
        wrapper.derrotas = 0
        return wrapper

    def contador_empates(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            resultado = func(*args, **kwargs)
            if resultado == "empate":
                wrapper.empates += 1
                print(f"Empates totales: {wrapper.empates}")
            return resultado
        wrapper.empates = 0
        return wrapper

    def limitar_partidas(max_partidas):
        def decorador(func):
            @wraps(func)
            def wrapper(*args, **kwargs):
                if wrapper.partidas < max_partidas:
                    wrapper.partidas += 1
                    return func(*args, **kwargs)
                else:
                    print(f"Has alcanzado el límite de {max_partidas} partidas por sesión.")
                    return None
            wrapper.partidas = 0
            return wrapper
        return decorador

    @registrar_tiempo
    @contador_juegos
    @registrar_fecha_hora
    @contador_victorias
    @contador_derrotas
    @contador_empates
    @limitar_partidas(10)
    def jugar_ahorcado():
        palabra = random.choice(palabras)
        letras_adivinadas = set()
        intentos = 6

        while intentos > 0:
            palabra_oculta = ''.join([letra if letra in letras_adivinadas else '_' for letra in palabra])
            print(f"\nPalabra: {palabra_oculta}")
            print(f"Intentos restantes: {intentos}")

            if palabra_oculta == palabra:
                print("¡Felicidades! Has adivinado la palabra.")
                return "victoria"

            letra = input("Ingresa una letra: ").lower()

            if letra in letras_adivinadas:
                print("Ya has intentado esa letra. Prueba otra.")
            elif letra in palabra:
                letras_adivinadas.add(letra)
                print("¡Correcto!")
            else:
                intentos -= 1
                print("Incorrecto. Pierdes un intento.")

        print(f"\n¡Has perdido! La palabra era: {palabra}")
        return "derrota"

    # Bucle principal del juego
    while True:
        jugar_ahorcado()
        if input("¿Quieres jugar otra vez? (s/n): ").lower() != 's':
            break

    print("¡Gracias por jugar!")

hangman()
```
